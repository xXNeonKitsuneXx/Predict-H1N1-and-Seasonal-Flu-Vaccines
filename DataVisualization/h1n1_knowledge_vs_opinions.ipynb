```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
training_features_path = '../training_set_features.csv'
training_features = pd.read_csv(training_features_path)
```

```python
# Handle columns with null values (drop as needed)
columns_with_nulls = ['h1n1_knowledge', 'opinion_h1n1_vacc_effective', 'opinion_h1n1_sick_from_vacc', 'age_group']
training_features = training_features[columns_with_nulls].dropna()
```

```python
# Group data by age group and calculate mean for vaccine opinions and knowledge
age_group_opinion_knowledge = training_features.groupby('age_group')[
    [
        'h1n1_knowledge',
        'opinion_h1n1_vacc_effective',
        'opinion_h1n1_sick_from_vacc',
    ]
].mean()
```

```python
# Plot the data
plt.figure(figsize=(12, 10))
```

```python
# Plot for H1N1 Knowledge
plt.subplot(2, 1, 1)
age_group_opinion_knowledge['h1n1_knowledge'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('H1N1 Knowledge by Age Group')
plt.ylabel('Average Knowledge Level')
plt.xlabel('Age Group')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Plot for H1N1 Knowledge
plt.subplot(2, 1, 1)
age_group_opinion_knowledge['h1n1_knowledge'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('H1N1 Knowledge by Age Group')
plt.ylabel('Average Knowledge Level')
plt.xlabel('Age Group')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Plot for Opinions about H1N1 Vaccine
plt.subplot(2, 1, 2)
age_group_opinion_knowledge[['opinion_h1n1_vacc_effective', 'opinion_h1n1_sick_from_vacc']].plot(
    kind='bar', figsize=(12, 8), color=['green', 'red'], edgecolor='black', ax=plt.gca()
)
plt.title('Opinions about H1N1 Vaccine by Age Group')
plt.ylabel('Average Opinion Level')
plt.xlabel('Age Group')
plt.xticks(rotation=45, ha='right')
plt.legend(['H1N1 Vaccine Effective', 'H1N1 Sick From Vaccine'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
