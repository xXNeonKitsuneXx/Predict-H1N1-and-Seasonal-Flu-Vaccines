```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
training_features_path = '../training_set_features.csv'
training_features = pd.read_csv(training_features_path)
```

```python
# Calculate the percentage of null/missing values for each column
missing_data = (training_features.isnull().sum() / len(training_features)) * 100
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
```

```python
# Plot the missing data as a bar chart
plt.figure(figsize=(12, 8))
missing_data.plot(kind='bar', color='gray', edgecolor='black')
plt.title('Percentage of Missing Data in Training Features')
plt.ylabel('Percentage of Missing Values (%)')
plt.xlabel('Columns')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
